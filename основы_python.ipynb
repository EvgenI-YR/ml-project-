#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(text):
    vowels = "aeiou"
    count = 0
    for i in text:
        if i in vowels:
            count += 1
    return count
def run_tests():
    assert count_vowels("apple") == 2, "Ошибка на слове apple"
    assert count_vowels("SKY") == 0, "Ошибка на слове SKY (заглавные)"
    assert count_vowels("aeiou") == 5, "Ошибка на всех гласных"
    print("УРАААААААА")
run_tests()

```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def pravda_losh(text):
    urfu = set()
    for i in text:
        if i in urfu:
            return False
        urfu.add(i)
    return True

assert pravda_losh("abc") == True, "hahaha"
assert pravda_losh("aaa") == False, "cring"
assert pravda_losh("") == True, "oooo"
assert pravda_losh("AbA") == False, "nice try"
assert pravda_losh("12345") == True, "lol kek"

print("закончили упражнение !")
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def dosta(n):
    count = 0
    while n >= 10 :
        okak = 1
        while n > 0 :
            n, ones = divmod(n, 10)
            okak *= ones
        n = okak
        count += 1
    return count
if __name__ == "__main__":
    assert dosta(4) == 0
    assert dosta(0) == 0
    assert dosta(7) == 0
    assert dosta(39) == 3
    assert dosta(999) == 4
print(' yes yes ye' )
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError("Векторы должны иметь одинаковую длину")
    n = len(pred)
    a = sum((p - t) ** 2 for p, t in zip(pred, true))
    return a / n

if __name__ == "__main__":
    assert mse([1, 2, 3], [1, 2, 3]) == 0.0
    assert mse([2, 3, 4], [1, 2, 3]) == 1.0  
print ('the end, HAPPY !!')
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
from typing import Tuple

def count_exponent(n: int, d: int) -> Tuple[int, int]:
    cnt = 0
    while n % d == 0:
        n //= d
        cnt += 1
    return cnt, n

def format_factor(p: int, e: int) -> str:
    return f"({p})" if e == 1 else f"({p}**{e})"

def prime_factorization(n: int) -> str:
    if n <= 1:
        return ""
    factors = []
    if n % 2 == 0:
        exp, n = count_exponent(n, 2)
        factors.append(format_factor(2, exp))
    p = 3
    while p * p <= n:
        if n % p == 0:
            exp, n = count_exponent(n, p)
            factors.append(format_factor(p, exp))
        p += 2
    if n > 1:
        factors.append(format_factor(n, 1))
    return "".join(factors)


# Тесты
assert prime_factorization(1) == ""
assert prime_factorization(2) == "(2)"
assert prime_factorization(4) == "(2**2)"
assert prime_factorization(6) == "(2)(3)"
assert prime_factorization(12) == "(2**2)(3)"
assert prime_factorization(86240) == "(2**5)(5)(7**2)(11)"
assert prime_factorization(999) == "(3**3)(37)"
# Для большого числа можно проверить, что функция не падает и возвращает непустую строку
result = prime_factorization(277777788888899)
assert isinstance(result, str) and len(result) > 0
print("Все тесты пройдены!")
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number: int):
    if number < 1:
        return "It is impossible"
    total = 0
    k = 0
    while total < number:
        k += 1
        total += k * k
        if total == number:
            return k
    return "It is impossible"

if __name__ == "__main__":
    assert pyramid(1) == 1
    assert pyramid(5) == 2
    assert pyramid(14) == 3
    assert pyramid(30) == 4
    assert pyramid(55) == 5
    assert pyramid(91) == 6
    assert pyramid(140) == 7
    assert pyramid(204) == 8
    assert pyramid(285) == 9
    assert pyramid(385) == 10
print('Все тесты пройдены!')
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(number: int) -> bool:
    digits = str(number)
    length = len(digits)
    if length % 2 == 1:
        mid = length // 2
        left_sum = sum(int(d) for d in digits[:mid])
        right_sum = sum(int(d) for d in digits[mid+1:])
    else:
        mid = length // 2
        left_sum = sum(int(d) for d in digits[:mid-1])
        right_sum = sum(int(d) for d in digits[mid+1:])
    return left_sum == right_sum

if __name__ == "__main__":
    assert is_balanced(1) is True
    assert is_balanced(5) is True
    assert is_balanced(9) is True
    assert is_balanced(101101) is True
    print("Все тесты пройдены!")
```
